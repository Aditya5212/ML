```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('iris.csv')
```

```python
target_data = df.iloc[:, 5]
```

```python
clustering_data = df.iloc[:, [1,2,3,4]]
clustering_data.head()
```

```python
correlation_matrix = clustering_data.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot = True, cmap='PuBu_r')
plt.title('CORRELATION MATRIX')
plt.show()
```

```python
df1 = df[df['Species']=='Iris-setosa']
df2 = df[df['Species']=='Iris-versicolor']
df3 = df[df['Species']=='Iris-virginica']

plt.scatter(df1['PetalLengthCm'],df1['PetalWidthCm'], color='steelblue' , label='Iris-setosa')
plt.scatter(df2['PetalLengthCm'],df2['PetalWidthCm'], color='lightblue', label='Iris-versicolor')
plt.scatter(df3['PetalLengthCm'],df3['PetalWidthCm'], color='navy' , label='Iris-virginica ')

plt.title('SCATTER PLOT ')
plt.legend()
plt.show()
```

```python

k_meansclus = range(1,11)
sse = []

for k in k_meansclus :
  km = KMeans(n_clusters =k)
  km.fit(clustering_data)
  sse.append(km.inertia_)
```

```python
plt.title('The Elbow Method')
plt.plot(k_meansclus,sse, marker= 'o')
plt.xlabel('Number Of Clusters')
plt.ylabel('sse')
plt.show()
```

```python
km1 = KMeans(n_clusters=3,max_iter=300 , random_state=0)
km1.fit(clustering_data)
```

```python
y_means = km1.fit_predict(clustering_data)
y_means
```

```python
km1.cluster_centers_
```

```python
# df_imp = np.array(clustering_data)
# df_imp
```

```python
plt.scatter(clustering_data.iloc[y_means == 0, 0], clustering_data.iloc[y_means == 0, 1], s=100, c='red', label='Iris-setosa')
plt.scatter(clustering_data.iloc[y_means == 1, 0], clustering_data.iloc[y_means == 1, 1], s=100, c='blue', label='Iris-versicolor')
plt.scatter(clustering_data.iloc[y_means == 2, 0], clustering_data.iloc[y_means == 2, 1], s=100, c='green', label='Iris-virginica')

# Plot the centroids of the clusters
plt.scatter(km1.cluster_centers_[:, 0], km1.cluster_centers_[:, 1], s=100, c='black', label='Centroids')
plt.title('K-Means Clustering of Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```
